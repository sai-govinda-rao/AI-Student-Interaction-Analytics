```python
df1 = df.copy()
```

```python
level_mapping = {
    "High School": 0,
    "Undergraduate": 1,
    "Graduate": 2
}

df1["StudentLevel"] = df["StudentLevel"].map(level_mapping)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
```

```python
num_cols = df1.select_dtypes(include="number").columns.to_list()
cat_cols = df1.select_dtypes(exclude='number').columns.tolist()

preprocessor = ColumnTransformer(
    transformers=[
        ("num", "passthrough", num_cols),
        ("cat", OneHotEncoder(drop="first", handle_unknown="ignore"), cat_cols)
    ],
    remainder="drop"
)
cat_cols
```

```python
target = df1["UsedAgain"]
df1 = df1.drop(["SessionID", "SessionDate", "UsedAgain"], axis=1)
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df1, target, test_size=0.2, random_state=42)
```

```python
x_train
```

### ML Model Building

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()

log_pipeline = Pipeline(steps=[
    ("preprocessing", preprocessor),
    ("model", log_reg)
])
log_pipeline.fit(x_train, y_train)
```

```python
log_pipeline.score(x_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dec_tree = DecisionTreeClassifier()

dec_tree_pipeline = Pipeline(steps=[
    ("preprocessing", preprocessor),
    ("model", dec_tree)
])
dec_tree_pipeline.fit(x_train, y_train)
```

```python
dec_tree_pipeline.score(x_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

rf_pipeline = Pipeline(steps=[
    ("preprocessing", preprocessor),
    ("model", rf_model)
])
rf_pipeline.fit(x_train, y_train)
```

```python
rf_pipeline.score(x_test, y_test)
```

```python
from sklearn.naive_bayes import GaussianNB

navie_base = GaussianNB()

navie_base_pipeline = Pipeline(steps=[
    ("preprocessing", preprocessor),
    ("model", navie_base)
])

navie_base_pipeline.fit(x_train, y_train)
```

```python
navie_base_pipeline.score(x_test, y_test)
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
param_grid_lr = {
    "model__C": [0.01, 0.1, 1, 10],
    "model__penalty": ["l1", "l2"],
    "model__solver": ["liblinear"]
}

grid_lr = grid_search = GridSearchCV(
        log_pipeline, 
        param_grid=param_grid_lr, 
        cv=cv, 
        scoring='f1', 
        n_jobs=-1
)

grid_lr.fit(x_train, y_train)
```

```python
grid_lr.score(x_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV


param_dist_dt = {
    "model__max_depth": [None, 5, 10, 20, 30],
    "model__min_samples_split": [2, 5, 10, 20],
    "model__min_samples_leaf": [1, 2, 5, 10],
    "model__criterion": ["gini", "entropy"]
}

random_dt = RandomizedSearchCV(
    dec_tree_pipeline,
    param_distributions=param_dist_dt,
    n_iter=30,
    cv=cv,
    scoring="f1",
    n_jobs=-1,
    random_state=42
)

random_dt.fit(x_train, y_train)
```

```python
random_dt.score(x_test, y_test)
```

```python
param_dist_rf = {
    "model__n_estimators": [100, 200, 500],
    "model__max_depth": [None, 10, 20, 30],
    "model__min_samples_split": [2, 5, 10],
    "model__min_samples_leaf": [1, 2, 4],
    "model__max_features": ["sqrt", "log2"]
}

random_rf = RandomizedSearchCV(
    rf_pipeline,
    param_dist_rf,
    n_iter=30,
    cv=cv,
    scoring="f1",
    n_jobs=-1,
    random_state=42
)

random_rf.fit(x_train, y_train)
```

```python
random_rf.score(x_test, y_test)
```

```python
param_grid_nb = {
    "model__var_smoothing": [1e-12, 1e-10, 1e-9, 1e-8, 1e-7]
}

grid_nb = GridSearchCV(
    navie_base_pipeline,
    param_grid=param_grid_nb,
    cv=cv,
    scoring="f1",   # or "f1"
    n_jobs=-1
)

grid_nb.fit(x_train, y_train)
```

```python
grid_nb.score(x_test, y_test)
```

```python
models = {
    "Logistic Regression": grid_lr,
    "Decision Tree": random_dt,
    "NavieBase": grid_nb,
    "Random Forest": random_rf
}

results = []

for name, model in models.items():
    results.append({
        "Model": name,
        "Best Score": model.best_score_,
        "Best Params": model.best_params_
    })

results_df = pd.DataFrame(results)
```
